```
import pandas as pd
df=pd.read_csv("EduData.csv")

df
```

```
#Normalization
from sklearn.preprocessing import MinMaxScaler

scaling=MinMaxScaler()
```

```
scaling.fit_transform(df[['raisedhands','Discussion']])
```

```
#Standardization
from sklearn.preprocessing import StandardScaler

stdscaling=StandardScaler()
```

```
stdscaling.fit_transform(df[['raisedhands','Discussion']])
```

```
#One Hot Encoding
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer 
import numpy as np

enc = ColumnTransformer( [('encoder', OneHotEncoder(), [0])], remainder='passthrough')
```

```
new = pd.DataFrame(data=enc.fit_transform(df),columns=['F', 'M',	'PlaceofBirth',	'StageID',	'GradeID', 'Topic'	,'Relation',	'raisedhands',	'VisITedResources'	,'AnnouncementsView','Discussion'	,'ParentAnsweringSurvey',	'ParentschoolSatisfaction',	'StudentAbsenceDays'])

new
```

```
#Imputing missing values
from sklearn.impute import KNNImputer

data=pd.read_csv("EduData.csv",usecols=['raisedhands','VisITedResources','AnnouncementsView','Discussion'])
data
```

```
missingdata=data.replace(0,np.nan)
imputer = KNNImputer(n_neighbors=10, weights="uniform")
arr = imputer.fit_transform(missingdata)
```

```
new = pd.DataFrame(data = arr, columns = ['raisedhands','VisITedResources','AnnouncementsView','Discussion'])
new
```
